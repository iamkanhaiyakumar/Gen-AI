```python
from abc import ABC, abstractmethod
```

```python
class Runnable(ABC):
    
    @abstractmethod
    def invoke(self, input_dict):
        pass
```

```python
import random
class nakliLLm(Runnable): #class

    def __init__(self): #constructor
        print('LLM created') #called constructor

    def invoke(self, prompt): #method
        respomse_list = [   #list of responses
            'Delhi is the capital of India',
            'IPL is a cricket league',
            'AI stands for Artificial Intelligence'
        ]
        return{'response': random.choice(respomse_list)} #returns a random response from the list


##  Alternate way to write the same code this is also correct but the above one is more concise and cleaner  
    # def predict(self, prompt): #method
    #     respomse_list = [   #list of responses
    #         'Delhi is the capital of India',
    #         'IPL is a cricket league',
    #         'AI stands for Artificial Intelligence'
    #     ]
    #     return{'response': random.choice(respomse_list)} #returns a random response from the list   
```

```python
class nakliPromptTemplate(Runnable): #class for prompt template
    
    def __init__(self, template, input_variables): #constructor for prompt template
        self.template = template #assigning template to self.template
        self.input_variables = input_variables #assigning input variables to self.input_variables

    def invoke(self, input_dict): #method to format the prompt template with input variables
        return self.template.format(**input_dict) #calling the format method

    def format(self, input_dict):   # method to format the prompt template with input variables
        return self.template.format(**input_dict)
    
    

## Alternate way to write the same code this is also correct but the above one is more concise and cleaner
    # def format(self, input_dict): #method to format the prompt template with input variables
    #     return self.template.format(**input_dict) #formatting the template with input variables using ** to unpack the dictionary
```

```python
class nakliStrOutputParser(Runnable):

    def __int__(self):
        pass

    def invoke(self, input_data):
        return input_data['response'] #returns the response from the input data
```

```python
class RunnableConnector(Runnable):

    def __init__(self, runnable_list): #constructor for RunnableConnector
        self.runnable_list = runnable_list #assigning runnable list to self.runnable_list

    def invoke(self, input_data):

        for runnable in self.runnable_list: #iterating through the runnable list
            input_data = runnable.invoke(input_data) #calling the invoke method of each runnable and updating the input data with the output of each runnable

        
        
        return input_data #returning the final output after passing through all the runnables
```

```python
template = nakliPromptTemplate( #creating an object of nakliPromptTemplate class
    template='write a {length} poem about {topic}?', #assigning template to the object
    input_variables=['length', 'topic']) #assigning template and input variables to the object
```

```python
llm = nakliLLm() #object creation
```

```python
parser = nakliStrOutputParser() #object creation
```

```python
chain = RunnableConnector([template,llm, parser])
```

```python
chain.invoke({'length': 'short', 'topic': 'nature'})
```
